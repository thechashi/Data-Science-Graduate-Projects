### Carnegie Mellon University

```python
table = S.find('div', {'id': 'Carnegie%20Mellon%20University-faculty'})
```

```python
df = pd.read_html(str(table))[0]
```

```python
df
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df = df.replace(to_replace='None', value=np.nan).dropna()
df.rename(columns={'Faculty':'Faculty Name'}, inplace=True)
df['University'] = 'Carnegie Mellon University'
df = df.reset_index()
df.drop('index', axis=1, inplace=True)
df
```

```python
headers = []
data = []
for row in table.find_all('tr'):
    row_data = []
    for cell in row.find_all(['td', 'th']):
        # Extract hyperlink if it exists
        link = cell.find('a')
        if link is not None:
            row_data.append(link.get('href'))
        else:
            row_data.append(cell.text.strip())
            
        # Extract header if it exists
        if cell.name == 'a':
            if link is not None:
                headers.append('link_row'+ row_data[-1])
    if row_data:
        data.append(row_data)

df2 = pd.DataFrame(data)
```

```python
df2
```

```python
df2.columns = df2.iloc[0]
df2 = df2[1:]
df2.rename(columns={'Faculty':'Faculty Home Page'}, inplace=True)
df2.rename(columns={'# Pubs':'publications'}, inplace=True)
```

```python
df2 = df2.replace(to_replace='None', value=np.nan).dropna()
```

```python
df2
```

```python
df2.to_csv("output.csv", index=False)
```

```python
df2 = df2.iloc[: , :-1]
df2 = df2.iloc[: , 1:]
df2.columns.values[1] = 'dblp'
df2 = df2.reset_index()
df2.drop('index', axis=1, inplace=True)
```

```python
df_car = pd.concat([df, df2], axis=1, join='inner')
print(df_car.columns)
df_car.columns = ["faculty_name", "publications", "adj", "university_name", "faculty_home_page", "dblp"]
df_car
```
